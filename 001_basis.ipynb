```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#定义数据
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
```

```python
#前向传播
def forward(x):
    return x * w
#定义损失函数
def loss(x, y):
    y_pred = forward(x)
    return 0.5 * (y_pred - y) **2
```

```python
w_list = []
mse_list = []
for w in np.arange(0.0, 4.1, 0.1):
    print("w=",w)
    L_sum = 0
    for x_val, y_val in zip(x_data, y_data):
        y_pred_val = forward(x_val)
        L = loss(x_val, y_val)
        L_sum += L
        print("\t", x_val, y_val, y_pred_val, L)
    print("MSE=", L_sum / 3)
    w_list.append(w)
    mse_list.append(L_sum / 3)
```

```python
#画图
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show()
```
